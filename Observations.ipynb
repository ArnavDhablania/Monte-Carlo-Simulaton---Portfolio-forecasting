# <u> Observations and Interpretations

## Objective

This study compares the **future performance of a growth-focused portfolio to the S&P 500 (VOO)**, with a main focus on **near-term outcomes through 2026**.

To better understand these results, the analysis also examines medium- and long-term horizons to see how the portfolio performs across different market cycles.

The goal is to understand when the portfolio tends to **perform well, when it does not, and how sensitive its results are to the time horizon**. The goal is not to claim guaranteed outperformance.

## Methodology

- Method: Monte Carlo simulation using historical return data

- Benchmark: S&P 500 (VOO)

- Simulations: ~5,000 paths per time horizon

- Time horizons analyzed:

  - Short-term: 3 years (focused on 2024â€“2026)

  - Medium-term: 10 years

  - Long-term: 20 years (full market-cycle view)

- Evaluation approach:

  - 5th percentile: realistic downside

  - 50th percentile: median outcome

  - 80th percentile: realistic upside

- Main comparison metric: Probability of outperforming the S&P 500 at the end of each horizon

## Section 3: Robustness and Synthesis

### <u> Robustness

- In this model, robustness is assessed by varying the historical calibration window (3-year, 10-year, and 20-year) used to estimate return dynamics.

- The portfolio exhibits strong performance advantages under shorter calibration windows, including higher expected returns, greater upside potential, and a higher probability of outperforming the VOO. Though as the calibration window increases, these advantages systematically weaken, indicating that excess performance is sensitive to recent market conditions rather than stable across long-term regimes.

### <u> Synthesis
- Regime Dependence and Convergence

    - Performance metrics consistently converged toward VOO as longer historical windows are incorporated.

    - Expected returns, Sharpe ratios, drawdowns, tail outcomes, and outperformance probabilities all become increasingly similar under the 20-year window.

    - This convergence reflects the inclusion of multiple market cycles, where periods of strong relative performance are offset by weaker regimes.

- Structural Context

    - VOO benefits from periodic index rebalancing and replacements, while the portfolio is held static throughout the forecast horizon.

    - This structural difference contributes to the benchmarkâ€™s long-run stability and limits the relative outperformance by the portfolio.

- Implications for Portfolio Stability

    - The portfolio demonstrates conditional robustness, performing strongly in favorable market conditions but converging toward benchmark-like behavior over longer horizons.

    - While not fragile, the portfolio does not consistently outperform in all market regimes, highlighting a trade-off between upside potential and long-run stability.
